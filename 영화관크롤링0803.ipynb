```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains 
import time
from bs4 import BeautifulSoup
import requests
import pandas as pd 
import numpy as np
pd.set_option('display.max_rows', None)
import re
```

```python
df = pd.read_csv("C:\\Users\yuna\Downloads\\Fixed_Movie_Data.csv")
df
```

```python
df = df.fillna("")
df
```

```python
df["New"] = df["poi_nm"] + df["branch_nm"]
df
```

```python
Movie_Places = df["New"]
Movie_Places
```

```python
path = "C:\\Users\\yuna\\OneDrive\\바탕 화면\\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.implicitly_wait(5)
```

```python
MovieId = [ ]
MoviePlaces = [ ]

for places in Movie_Places:
    driver.get("https://map.naver.com/v5/search")
    driver.find_element_by_css_selector("div.input_box>input.input_search").send_keys(places + "\n")
    element = driver.find_element_by_id("searchIframe")
    driver.switch_to.frame(element)
    time.sleep(5)  
    if driver.find_elements_by_css_selector("div div._1--tV"):
        MoviePlaces.append(places + " " + "없음")
        MovieId.append(" ")
    else:                
        try:
            if driver.find_elements_by_css_selector("div span._3Apve"):
                posts = driver.find_elements_by_css_selector("div span._3Apve")
                posts[0].click()
                driver.switch_to.default_content() 
                html = driver.page_source 
                soup = BeautifulSoup(html,"html.parser")   # 드라이버 현재 페이지의 html 정보 가져오기 
                current_url = driver.current_url            
                reg = re.compile('place/([0-9]+)?')
                naver_id = reg.findall(current_url)[0]  
                MovieId.append(naver_id)
                MoviePlaces.append(places)     
    
 
                element = driver.find_element_by_id("searchIframe")
                driver.switch_to.frame(element)
                time.sleep(5)
                
            elif driver.find_elements_by_css_selector("div span._1AjbI"):
                posts = driver.find_elements_by_css_selector("div span._1AjbI")
                posts[0].click()
                driver.switch_to.default_content() 
                html = driver.page_source 
                soup = BeautifulSoup(html,"html.parser")   # 드라이버 현재 페이지의 html 정보 가져오기 
                current_url = driver.current_url            
                reg = re.compile('place/([0-9]+)?')
                naver_id = reg.findall(current_url)[0]#          
                MovieId.append(naver_id)
                MoviePlaces.append(places)


                
            else:
                html = driver.page_source 
                soup = BeautifulSoup(html,"html.parser")   # 드라이버 현재 페이지의 html 정보 가져오기 
                current_url = driver.current_url            
                reg = re.compile('place/([0-9]+)?')
                naver_id = reg.findall(current_url)[0]
                MovieId.append(naver_id)
                MoviePlaces.append(places)

                
              
        except:
            html = driver.page_source 
            soup = BeautifulSoup(html,"html.parser")   # 드라이버 현재 페이지의 html 정보 가져오기 
            current_url = driver.current_url            
            reg = re.compile('place/([0-9]+)?')

            naver_id = reg.findall(current_url)[0]
            MovieId.append(naver_id)
            MoviePlaces.append(places)
# 
```

```python
df["ID"] = pd.DataFrame(MovieId, columns = ["ID"])
df["PLACES"] = pd.DataFrame(MoviePlaces, columns = ["PLACES"])
df
```
