```python
import pandas as pd 
import numpy as np 


df1 = pd.read_csv("C:\\Users\\yuna\\OneDrive\\Documents\\카카오톡 받은 파일\\뉴런장소.csv",encoding='CP949')
df1
```

```python
df2 = pd.read_csv("C:\\Users\\yuna\\OneDrive\\Documents\\카카오톡 받은 파일\\국내영화관(한국문화정보원).csv")
df2
```

```python
df2 = df2.replace(np.nan,"",regex = True)
df2
```

```python
df2 = df2[df2.mcate_nm != "비디오/DVD감상실"]
df2
```

```python
df2["New"] = df2["poi_nm"] + " " + df2["branch_nm"]
df2
```

```python
df11 = df1[["STORE_SN","STORE_NM"]]
df11
```

```python
df11_CGV_LOTTE = df11.loc[(df11["STORE_NM"].str.startswith("CGV"))|(df11["STORE_NM"].str.startswith("롯데시네마"))]
df11_CGV_LOTTE
```

```python
df2_CGV_LOTTE = df2.loc[(df2["New"].str.startswith("CGV"))|(df2["New"].str.startswith("롯데시네마"))]
df2_CGV_LOTTE
```

```python
CGV_LOTTE_DUPLITCATED = pd.merge(df11_CGV_LOTTE,df2_CGV_LOTTE, left_on = "STORE_NM", right_on = "New", how = "inner")
CGV_LOTTE_DUPLITCATED # CGV랑 롯데는 공통점이없다
```

```python
inner_merge = pd.merge(df11,df2,left_on= "STORE_NM", right_on = "poi_nm", how = "inner")
inner_merge
```

```python
right_merge = pd.merge(df11,df2,left_on = "STORE_NM", right_on = "poi_nm", how = "right")
right_merge
```

```python
Final_df = right_merge.loc[right_merge["STORE_NM"] != right_merge["poi_nm"]]
Final_df
```

```python
Fixed_Movie_Data = Final_df.drop(["STORE_SN","STORE_NM","New"],axis= 1)
Fixed_Movie_Data
```

```python
Fixed_Movie_Data.to_csv("C:\\Users\\yuna\\Downloads\\Fixed_Movie_Data.csv", index = None)
```
